```python
import pandas as pd
import altair as alt
import numpy as np
import matplotlib.pyplot as plt
from vega_datasets import data
```

```python
df = pd.read_csv("sunshine.csv")
df
```

```python
# Subset the data to only include the month of April
df_apr = df[df["month"] == "Apr"]
df_apr
```

```python
states = alt.topo_feature(data.us_10m.url, "states")
background = alt.Chart(states).mark_geoshape(
    fill="lightgray",
    stroke="white"
).project("albersUsa").properties(
    width=500,
    height=400
)

cities = alt.Chart(df_apr).mark_circle(
    size=200,  # Fixed size
    opacity=0.7,
    stroke="black",
    strokeWidth=1
).encode(
    longitude='lon:Q',
    latitude='lat:Q',
    color=alt.Color('sunshine:Q', scale=alt.Scale(scheme='oranges'), title="Sunshine Hours"),
    tooltip=['city:N', 'sunshine:Q']
).project("albersUsa").properties(
    width=500,
    height=400
)

background + cities
```
